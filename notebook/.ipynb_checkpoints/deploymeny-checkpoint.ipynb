```python
from huggingface_hub import snapshot_download

repo_path = snapshot_download(repo_id=MODEL_PERSONAL_NAME)

pipeline = CatVTONPipeline(
    base_ckpt=MODEL_NAME,
    attn_ckpt=repo_path,
    attn_ckpt_version="mix",
    weight_dtype=init_weight_dtype("bf16"),
    use_tf32=True,
    device='cuda'
)

mask_processor = VaeImageProcessor(
    vae_scale_factor=8, 
    do_normalize=False, 
    do_binarize=True, 
    do_convert_grayscale=True
)

automasker = AutoMasker(
    densepose_ckpt=os.path.join(repo_path, "DensePose"),
    schp_ckpt=os.path.join(repo_path, "SCHP"),
    device='cuda'
)
```

```python
import base64
import io
from PIL import Image
import torch

def submit_function_b64(
    person_image_b64: str,
    cloth_image_b64: str,
    cloth_type: str,
    num_inference_steps: int,
    guidance_scale: float,
    seed: int = -1,
    width: int = 768,
    height: int = 1024,
):
    """
    Args
    ----
    person_image_b64 : str
        Base-64-encoded RGB image of the person.
    cloth_image_b64  : str
        Base-64-encoded RGB image of the garment.
    cloth_type       : str
        Category used by `automasker` (e.g. "upper", "lower").
    num_inference_steps : int
        Diffusion steps for the pipeline.
    guidance_scale   : float
        CFG strength.
    seed             : int
        >-1 locks stochasticity; −1 keeps it random.
    width, height    : int
        Target resolution fed to the pipeline.

    Returns
    -------
    PIL.Image.Image
        Synthesised try-on image.
    """

    # ------------------------------------------------------------------
    # 1. Decode base-64 → PIL
    # ------------------------------------------------------------------
    def b64_to_pil(b64_str: str) -> Image.Image:
        return Image.open(io.BytesIO(base64.b64decode(b64_str))).convert("RGB")

    person_image = b64_to_pil(person_image_b64)
    cloth_image  = b64_to_pil(cloth_image_b64)

    # ------------------------------------------------------------------
    # 2. Pre-process
    # ------------------------------------------------------------------
    person_image = resize_and_crop(person_image, (width, height))
    cloth_image  = resize_and_padding(cloth_image, (width, height))

    # ------------------------------------------------------------------
    # 3. Optional seeding
    # ------------------------------------------------------------------
    generator = None
    if seed != -1:
        generator = torch.Generator(device="cuda").manual_seed(seed)

    # ------------------------------------------------------------------
    # 4. Build mask
    # ------------------------------------------------------------------
    mask = automasker(person_image, cloth_type)["mask"]
    mask = mask_processor.blur(mask, blur_factor=9)

    # ------------------------------------------------------------------
    # 5. Run diffusion pipeline
    # ------------------------------------------------------------------
    result_image = pipeline(
        image=person_image,
        condition_image=cloth_image,
        mask=mask,
        num_inference_steps=num_inference_steps,
        guidance_scale=guidance_scale,
        generator=generator,
    )[0]

    return result_image
```

```python
def file_to_b64(path):
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

person_b64 = file_to_b64("../sample_images/person_image.jpg")
cloth_b64  = file_to_b64("../sample_images/dress.jpg")

output_img = submit_function_b64(
    person_image_b64 = person_b64,
    cloth_image_b64  = cloth_b64,
    cloth_type       = "upper",
    num_inference_steps = 50,
    guidance_scale   = 2.5,
    seed             = 42,
)
```

```python
"""
PyFunc wrapper for CatVTON virtual-try-on (base-64 I/O).
"""

from __future__ import annotations
import base64, logging, os, shutil
from io import BytesIO
from pathlib import Path
from typing import Any, Dict, Tuple, List

import mlflow
import pandas as pd
import torch
from PIL import Image
from mlflow.models.signature import ModelSignature
from mlflow.types import ColSpec, ParamSchema, ParamSpec, Schema


logger = logging.getLogger(__name__)


class VirtualTryOnModel(mlflow.pyfunc.PythonModel):
    # ─────────────────────────── helpers ─────────────────────────────────────
    @staticmethod
    def _b64_to_pil(b64_str: str) -> Image.Image:
        if b64_str.startswith("data:"):
            b64_str = b64_str.split(",", 1)[1]
        return Image.open(BytesIO(base64.b64decode(b64_str))).convert("RGB")

    def _png_b64(self, img: Image.Image) -> str:
        buf = BytesIO()
        img.save(buf, format="PNG")
        return base64.b64encode(buf.getvalue()).decode()

    # ─────────────────────────── preprocessing ───────────────────────────────
    def _preprocess(
        self, row: pd.Series
    ) -> Tuple[Image.Image, Image.Image, Dict[str, Any]]:
        person = self._b64_to_pil(row["person_image"])
        cloth  = self._b64_to_pil(row["cloth_image"])

        meta = dict(
            cloth_type          = row.get("cloth_type", "upper"),
            num_inference_steps = int(row.get("num_inference_steps", 50)),
            guidance_scale      = float(row.get("guidance_scale", 7.5)),
            seed                = int(row.get("seed", -1)),
            width               = int(row.get("width", 768)),
            height              = int(row.get("height", 1024)),
        )

        person = resize_and_crop(person, (meta["width"], meta["height"]))
        cloth  = resize_and_padding(cloth, (meta["width"], meta["height"]))

        return person, cloth, meta

    # ─────────────────────────── load_context ────────────────────────────────
    def load_context(self, context: mlflow.pyfunc.PythonModelContext):
        """
        Instantiates CatVTONPipeline + helpers from the run’s artifacts.
        Expected artifact layout (set by `log_model`):
            artifacts/
                base_ckpt/      (Stable-Diffusion weights)
                attn_repo/      (snapshot_download folder)
        """
        base_ckpt = context.artifacts["base_ckpt"]
        attn_repo = context.artifacts["attn_repo"]

        self.pipe: CatVTONPipeline = CatVTONPipeline(
            base_ckpt = base_ckpt,
            attn_ckpt = attn_repo,
            attn_ckpt_version = "mix",
            weight_dtype = init_weight_dtype("bf16"),
            use_tf32 = True,
            device = "cuda" if torch.cuda.is_available() else "cpu",
        )

        # helpers
        self.mask_processor = VaeImageProcessor(
            vae_scale_factor = 8,
            do_normalize = False,
            do_binarize = True,
            do_convert_grayscale = True,
        )
        self.automasker = AutoMasker(
            densepose_ckpt = os.path.join(attn_repo, "DensePose"),
            schp_ckpt      = os.path.join(attn_repo, "SCHP"),
            device         = self.pipe.device,
        )
        logger.info("CatVTON pipeline & helpers loaded")

    # ─────────────────────────── predict ─────────────────────────────────────
    def predict(
        self,
        context: mlflow.pyfunc.PythonModelContext,
        model_input: pd.DataFrame,
        params: Dict[str, Any] | None = None,
    ) -> pd.DataFrame:
        params = params or {}
        return_mask: bool = bool(params.get("return_mask", False))

        outputs: List[Dict[str, Any]] = []
        for _, row in model_input.iterrows():
            try:
                person, cloth, meta = self._preprocess(row)

                mask = self.automasker(person, meta["cloth_type"])["mask"]
                mask = self.mask_processor.blur(mask, blur_factor=9)

                gen = (
                    torch.Generator(device=self.pipe.device).manual_seed(meta["seed"])
                    if meta["seed"] != -1
                    else None
                )

                result = self.pipe(
                    image = person,
                    condition_image = cloth,
                    mask = mask,
                    num_inference_steps = meta["num_inference_steps"],
                    guidance_scale = meta["guidance_scale"],
                    generator = gen,
                )[0]

                row_out = {
                    "result_image": self._png_b64(result),
                    "status": "ok",
                }
                if return_mask:
                    row_out["mask_image"] = self._png_b64(mask)
                outputs.append(row_out)
            except Exception as e:
                logger.exception("Inference failed")
                outputs.append(
                    {"result_image": None, "mask_image": None, "status": f"error: {e}"}
                )
        return pd.DataFrame(outputs)

    # ─────────────────────────── log_model helper ────────────────────────────
    @classmethod
    def log_model(
        cls,
        *,
        artifact_path: str,
        base_ckpt_dir: str | Path,
        attn_repo_dir: str | Path,
        demo_folder: str | Path | None = None,
        extra_pip: list[str] | None = None,
    ):
        """
        Call inside an active mlflow.start_run().
        """
        base_ckpt_dir = Path(base_ckpt_dir)
        attn_repo_dir = Path(attn_repo_dir)
        if not base_ckpt_dir.is_dir() or not attn_repo_dir.is_dir():
            raise ValueError("Checkpoint dirs not found")

        # IO schema ----------------------------------------------------------
        in_schema = Schema(
            [
                ColSpec("string", "person_image"),
                ColSpec("string", "cloth_image"),
                ColSpec("string", "cloth_type"),
                ColSpec("long", "num_inference_steps"),
                ColSpec("double", "guidance_scale"),
                ColSpec("long", "seed"),
                ColSpec("long", "width"),
                ColSpec("long", "height"),
            ]
        )
        out_schema = Schema(
            [
                ColSpec("string", "result_image"),
                ColSpec("string", "status"),
                ColSpec("string", "mask_image"),
            ]
        )
        param_schema = ParamSchema([ParamSpec("return_mask", "boolean", False)])
        sig = ModelSignature(inputs=in_schema, outputs=out_schema, params=param_schema)

        # requirements -------------------------------------------------------
        reqs = [
            "mlflow",
            "torch>=2.2.0",
            "huggingface_hub>=0.22.2",
            "catvton",                 # your package – adjust if different
            "Pillow",
            "accelerate",
        ]
        if extra_pip:
            reqs.extend(extra_pip)

        artifacts = {
            "base_ckpt": str(base_ckpt_dir),
            "attn_repo": str(attn_repo_dir),
        }
        if demo_folder:
            artifacts["demo"] = str(demo_folder)

        mlflow.pyfunc.log_model(
            artifact_path = artifact_path,
            python_model  = cls(),
            artifacts     = artifacts,
            pip_requirements = reqs,
            signature     = sig,
        )
        logger.info("CatVTON pyfunc logged at %s", artifact_path)
```
